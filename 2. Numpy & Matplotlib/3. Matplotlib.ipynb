# Basic Plotting

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2, 4, 6, 4])
plt.ylabel('Numbers')
plt.xlabel('Indices')
plt.title('MyPlot')
plt.show()
```

```python
# First List: X Axis
# Second List: Y Axis
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid()
plt.show()
```

```python
import numpy as np
t = np.arange(0., 5., 0.2)

# Blue Dashes, Red Squares, Green Triangles
plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t, t**2.2, 'rs', label='^2.2')
plt.plot(t, t**2.5, 'g^', label='^2.5')
plt.grid()

# Add Legend: Show which color is represented by which lines 
plt.show() 
```

# Controlling Line Properties

```python
x = [1, 2, 3, 4]
y = [5, 6, 7, 8]
plt.plot(x, y, linewidth=5.0)
plt.show()
```

### Using setp(): setproperties

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [2, 4, 6, 8]

lines = plt.plot(x1, y1, x2, y2)

plt.setp(lines[0], color='r', linewidth=2.0)
plt.setp(lines[1], color='g', linewidth=2.0)

plt.grid()
```

# Working With Multiple Figures & Axes

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

# Plot 1 Figure
plt.figure(1)

# Number of rows, Number of columns, Figure number
plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), 'b-')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
# First figure
plt.figure(1)
plt.subplot(211)
plt.plot([1, 2, 3])
plt.grid()

plt.subplot(212)
plt.plot([4, 5, 6])
plt.grid()

# Making a new figure 2
plt.figure(2)
plt.plot([4, 5, 6])
plt.grid()

# Going back to figure 1 again
plt.figure(1)
plt.subplot(211)
plt.title('Easy as 1, 2, 3')
plt.show()
```
