```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('iris.csv')
df_setosa = df[df['species'] == 'setosa']
df_virginica = df[df['species'] == 'virginica']
df_versicolor = df[df['species'] == 'versicolor']
```

## Box Plots & Whiskers

* Can easily understand percentile values from box plots

```python
sns.boxplot(x='species', y='petal_length', data=df)
plt.show()
```

<img src='box_plot_1.png'>

* If we declare a flower to be versicolor when petal_length < 5, then we are wrong (25% error) because 25th percentile of virginica is at 5 meaning that 25% of virginica flowers have petal length < 5
* The straight lines above and below are known as whiskers
* There is no standard way of drawing whiskers
    - Can take min & max values
    - seaborn takes 1.5 * IQR (generally)
* Whiskers are drawn in such a way that most of the points lie in the whiskers

## Violin Plots

* Already internally has box plots
* The white dot is 50th percentile
* Above and below are 25th & 75th percentile
* The lines above and below 25th & 75th percentile are whiskers
* Violin plots, plots the pdf at the sides

```python
sns.violinplot(x='species', y='petal_length', data=df)
plt.grid()
plt.show()
```

## Contour Plots

* Can visualize 2D density plots

```python
sns.jointplot(x='petal_length', y='petal_width', data=df_setosa, kind='kde')
plt.grid()
plt.show()
```
